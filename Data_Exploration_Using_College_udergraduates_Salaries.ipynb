```python
# The Highest Mid-Career Salary
print(clean_df['Mid-Career Median Salary'].max())
print(f"Index of the median salary is {clean_df['Mid-Career Median Salary'].idxmax()}")
print(clean_df['Undergraduate Major'][8])
```

```python
# Finding the difference between 10th and 90th Percentile Salary
sub_val = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1, 'Spread', sub_val)
clean_df.head()
```

```python
# Sorting The lowest Spread
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Sorting The Highest Top 5 degres according 90th percentile salary
high_risk = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_risk[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Finding the difference between Starting Mid-Salary & Median Mid-Salary
diff_sal = clean_df['Mid-Career Median Salary'].subtract(clean_df['Starting Median Salary'])
#clean_df.insert(2, 'Spread (diff_sal)', diff_sal)
clean_df.head()
```

```python
high_risk_sal = clean_df.sort_values('Spread (diff_sal)', ascending=False)
high_risk_sal[["Undergraduate Major", "Spread (diff_sal)"]].head()
```

```python
# The Highest Earning Majors
high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# The High Earners With Mid-Career Salary
high_risk=clean_df.sort_values('Mid-Career Median Salary', ascending=False)
high_risk[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

```python
# Defining the Number Of Major Groups using groupby()
clean_df.groupby("Group").count()
```

```python
 # Defining The Average Salary According to The Group
clean_df.groupby('Group').mean(numeric_only=True)
```

```python
# Changing Float Format 
pn.options.display.float_format = "{:,.2f}" .format
clean_df.groupby("Group").mean(numeric_only=True)
```
